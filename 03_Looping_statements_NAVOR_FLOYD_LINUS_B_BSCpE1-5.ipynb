```python
# Lab 2 Exercise 1
# Input an integer, 'prime' if the integer is prime else display 'not prime'.

while True:
    try:
        n = int(input("Enter an integer: "))
    except ValueError:
        print("Invalid input. Please enter an integer.")
        continue

    if n < 2:
        print("not prime")
    else:
        for i in range(2, (n // 2) + 1):
            if n % i == 0:
                print("not prime")
                break
        else:
            print("prime")

    choice = input("Do you want to check another number? (y/n): ").strip().lower()
    if choice == 'y':
        continue
    elif choice == 'n':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid input. Exiting by default.")
```

```python
# Lab 2 Exercise 2
# Compute for the cube of integers
while True:
    try:
        size = int(input("Enter the size of the array: "))
        elements = list(map(int, input("Enter the elements separated by space: ").split()))

        if len(elements) != size:
            print("Error: Number of elements does not match the size.")
        else:
            for num in elements:
                print(num**3)

    except ValueError:
        print("Invalid input. Please enter integers only.")

    choice = input("Do you want to check another number? (y/n): ").strip().lower()
    if choice == 'y':
        continue
    elif choice == 'n':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid input. Exiting by default.")
```

```python
# Lab 2 Exercise 3
# Generate a hollow square + .
while True:
    try:
        n = int(input("Enter the side length: "))

        if n > 1:
            for i in range(n):
                for j in range(n):
                    if i == 0 or i == n-1 or j == 0 or j == n-1:
                        print("+", end=" ")
                    else:
                        print(" ", end=" ")
                print()
        else:
            print("+")

    except ValueError:
        print("Invalid input. Please enter an integer.")

    choice = input("Do you want to check another number? (y/n): ").strip().lower()
    if choice == 'y':
        continue
    elif choice == 'n':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid input. Exiting by default.")
```

```python
# Lab 2 Exercise 4
# inverted right triangle

while True:
    try:
        n = int(input("Enter the height: "))

        if n > 0:
            for i in range(n, 0, -1):
                for j in range(i):
                    print("*", end=" ")
                print()
        else:
            print("Please enter a positive integer.")

    except ValueError:
        print("Invalid input. Please enter an integer.")

    choice = input("Do you want to check another number? (y/n): ").strip().lower()
    if choice == 'y':
        continue
    elif choice == 'n':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid input. Exiting by default.")
```
